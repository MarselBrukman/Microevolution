```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#s from 0 to 0.02
ss = np.linspace(0,0.02,100)
N = 100
x = []
y = []
for s in ss:
    a = 0
    b = 0
    for i in range(1000):
        popul = np.zeros(N,dtype=int)
        popul[0] = 1
        while np.sum(popul) !=0 and np.sum(popul) !=N:
            pr = [float(i) for i in popul]
            n_1=len(np.where(popul==1)[0])
            n_0=len(np.where(popul==0)[0])
            prob1=1/(n_1+(1-s)*n_0)
            prob0=(1-s)/(n_1+(1-s)*n_0)
            for j in np.where(popul==1)[0]:
                pr[j]=prob1
            for h in np.where(popul==0)[0]:
                pr[h]=prob0
            popul = np.random.choice(popul,(N,),p=pr)
        if np.sum(popul) == 0:
            a+=1
        else:
            b+=1
    y.append(b/1000)
    x.append(s)
```

```python
z = [0,0.02]
e = [1/N,1/N]
t = [0,0.02]
w = [0,0.04]

plt.plot(x,y)
plt.plot(z,e)
plt.plot(t,w)
```
